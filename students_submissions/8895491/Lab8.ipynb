Calculate the dot product between two vectors each of length 3 

```python
import torch


# calculating the dot product between two vectors `p_vc1` and `p_vc2` using the `torch.dot()` function. 
p_vc1 = torch.tensor([5, 6, 7])
p_vc2 = torch.tensor([9, 10, 11])

p_dot_ans = torch.dot(p_vc1, p_vc2)
print(f"Dot product of p_vc1 and p_vc2:\n{p_dot_ans}")
```

Calculate the matrix multiplication of two 2-dimensional matrices, each with a size of (2,2)

```python
# performing matrix multiplication between two 2-dimensional tensors `p2d_vc1` and `p2d_vc2`.
p2d_vc1 = torch.tensor([[5, 6], [5, 9]])
p2d_vc2 = torch.tensor([[9, 6], [4, 8]])

p_mult_ans = torch.matmul(p2d_vc1, p2d_vc2)
print(f"Multiplication of p2d_vc1 and p2d_vc2: \n{p_mult_ans}")
```
